```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
from datasets import load_dataset

dataset = load_dataset("comHannah/bokeh-dataset")
```

```python
bokeh_images = dataset["train"][:1351]["bokeh_image"]
```

```python
destination_folder = "output/dataset/bokeh_images/"

os.makedirs(destination_folder, exist_ok=True)
```

```python
for i, image in enumerate(bokeh_images):

    try:
        filename = os.path.basename(image.numpy()) 
    except (AttributeError, ValueError): 
        filename = f"image_{i+1}.jpg" 

    destination_path = os.path.join(destination_folder, filename)
    opencv_image = np.array(image)  
    opencv_image = cv2.cvtColor(opencv_image, cv2.COLOR_RGB2BGR) 
    cv2.imwrite(destination_path, opencv_image)

    print(f"Image saved successfully: {destination_path}")
```

```python
motion_blur_path = "/kaggle/working/output/dataset/motion_blur_path_new/"
defocused_blur_path = "/kaggle/working/output/dataset/defocused_blur_path_new/"


os.makedirs(motion_blur_path, exist_ok=True)  # Safe folder creation
os.makedirs(defocused_blur_path, exist_ok=True)  # Safe folder creation
```

```python
dataset_path = "/kaggle/input/bokeh-images-1/kaggle/working/output/dataset/bokeh_images"
```

```python
import random
#To choose the range of blur
def random_odd_number():
    number = random.randint(21, 31)
    while number % 2 == 0: 
        number = random.randint(21, 31) 
    return number
```

```python
def motion_blur(image_path, folder_path):
    # Load the image
    image = cv2.imread(image_path)

    # Kernel size for motion blur
    ganjil = random_odd_number()
    
    kernel_size = ganjil + 3
    kernel_h = np.zeros((kernel_size, kernel_size))
    kernel_h[int((kernel_size - 1)/2), :] = np.ones(kernel_size) 
    kernel_h /= kernel_size 


    horizontal_mb = cv2.filter2D(image, -1, kernel_h) 
    
    new_filename = os.path.join(folder_path, os.path.basename(image_path))

    if cv2.imwrite(new_filename, horizontal_mb):
        print(f"Image saved successfully: {new_filename}")
    else:
        print("Error! Failed to save image.")

def defocused_blur(image_path, folder_path):

    image = cv2.imread(image_path)
    
    ganjil = random_odd_number()
    
    kernel_size = (ganjil, ganjil)

    defocused_blurred_image = cv2.GaussianBlur(image, kernel_size, 0)

    new_filename = os.path.join(folder_path, os.path.basename(image_path))

    if cv2.imwrite(new_filename, defocused_blurred_image):
        print(f"Image saved successfully: {new_filename}")
    else:
        print("Error! Failed to save image.")
```

```python
import matplotlib.pyplot as plt
import random

#Test the blur parameter (kernel size)
image = cv2.imread("/kaggle/input/bokeh-images-1/kaggle/working/output/dataset/bokeh_images/image_73.jpg")

kernel_size = (21, 21)
new_filename = "image_73.jpg"
# Apply Gaussian blur
defocused_blurred_image = cv2.GaussianBlur(image, kernel_size, 0)
cv2.imwrite(new_filename, defocused_blurred_image)

# Display the blurred image
plt.imshow(cv2.cvtColor(defocused_blurred_image, cv2.COLOR_BGR2RGB))
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Defocused Blurred Image')
plt.show()
```

```python
for img_name in os.listdir(dataset_path):
    img_path = os.path.join(dataset_path, img_name)

    motion_blur(img_path, motion_blur_path)

    defocused_blur(img_path, defocused_blur_path)
```
